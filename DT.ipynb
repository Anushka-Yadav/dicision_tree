```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
iris.data
```

```python
x=iris.data
```

```python
y=iris.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dicision_tree_clf = DecisionTreeClassifier(criterion = "entropy")
```

```python
dicision_tree_clf.fit(x_train,y_train)
```

```python
y_pred=dicision_tree_clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import plot_tree
```

```python
from matplotlib.pylab import rcParams
rcParams["figure.figsize"]=80,50
plot_tree(dicision_tree_clf)
```

```python
dicision_tree_clf.predict([[5.1,3.5,1.4,0.2]])
```

```python
def predict(array,clf):
    val=clf.predict(array)[0]
    if val==0:
        return 'setosa'
    elif val==1:
        return 'versicolor'
    else:
        return 'verginica'

```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_clf = KNeighborsClassifier(n_neighbors=3)
```

```python
knn_clf.fit(x_train,y_train)
```

```python
y_pred_knn=knn_clf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_knn)
```

```python
for i in range(3,12,2):
    knn_clf = KNeighborsClassifier(n_neighbors=i)
    knn_clf.fit(x_train,y_train)
    y_pred_knn = knn_clf.predict(x_test)
    print(i,accuracy_score(y_test,y_pred_knn))
```

```python
predict([[5.1,3.5,1.4,0.2]],knn_clf)
```
